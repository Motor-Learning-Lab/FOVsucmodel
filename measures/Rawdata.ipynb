```python

def process_point_data(input_csv, output_csv):
    # Step 1: Read the input CSV file
    df = pd.read_csv(input_csv)
    
    # Step 2: Select specific columns
    # Assuming the columns you want to select are 'column1', 'column2', 'column3'
    selected_columns = ['Xr', 'Yr', 'date','trial_id','fish_id','session']
    df_selected = df[selected_columns]
    
    # Step 3: Initialize new columns
    df_selected['trial'] = 1
    df_selected['ds'] = 0.0
    df_selected['Yds'] = 0.0
    df_selected['dus'] = 0.0
    df_selected['duds'] = 0.0

    # Step 4: Calculate the 'trial' column
    for i in range(1, len(df_selected)):
        if df_selected['trial_id'].iloc[i] == df_selected['trial_id'].iloc[i-1]:
            df_selected.at[i, 'trial'] = df_selected['trial'].iloc[i-1]
        else:
            df_selected.at[i, 'trial'] = df_selected['trial'].iloc[i-1] + 1

    # Step 5: Calculate the 'ds', 'Yds', and 'dus' columns
    for i in range(1, len(df_selected)):
        ds_step = math.sqrt((df_selected['Xr'].iloc[i] - df_selected['Xr'].iloc[i-1])**2 + (df_selected['Yr'].iloc[i] - df_selected['Yr'].iloc[i-1])**2)
        Yds_step = df_selected['Yr'].iloc[i] * ds_step
        du_step = ((target_x - df_selected['Xr'].iloc[i-1]) * (df_selected['Xr'].iloc[i] - df_selected['Xr'].iloc[i-1]) + 
                   (target_y - df_selected['Yr'].iloc[i-1]) * (df_selected['Yr'].iloc[i] - df_selected['Yr'].iloc[i-1])) / ds_step
        duds_step = ((target_x - df_selected['Xr'].iloc[i-1]) * (df_selected['Xr'].iloc[i] - df_selected['Xr'].iloc[i-1]) + 
                   (target_y - df_selected['Yr'].iloc[i-1]) * (df_selected['Yr'].iloc[i] - df_selected['Yr'].iloc[i-1]))*ds_step

        if df_selected['trial'].iloc[i] == df_selected['trial'].iloc[i-1]:
            df_selected.at[i, 'ds'] = df_selected['ds'].iloc[i-1] + ds_step
            df_selected.at[i, 'Yds'] = df_selected['Yds'].iloc[i-1] + Yds_step
            df_selected.at[i, 'dus'] = df_selected['dus'].iloc[i-1] + du_step
            df_selected.at[i, 'duds'] = df_selected['duds'].iloc[i-1] + duds_step            
        else:
            df_selected.at[i, 'ds'] = 0
            df_selected.at[i, 'Yds'] = 0
            df_selected.at[i, 'dus'] = 0
            df_selected.at[i, 'duds'] = 0
    
    # Step 5: Save the new DataFrame to a new CSV file
    df_selected.to_csv(output_csv, index=False)
```

```python
def extract_trial_changes(input_csv, output_csv):
    # Step 1: Read the input CSV file
    df = pd.read_csv(input_csv)
    
    # Step 2: Select specific columns
    selected_columns = ['Xr', 'Yr', 'date', 'trial', 'fish_id', 'session','ds','Yds','dus','duds']
    df_selected = df[selected_columns]
    
    # Step 3: Initialize a list to collect rows where trial_id changes
    changed_rows = []
    
    # Step 4: Iterate over rows to identify changes in trial_id
    for i in range(1, len(df_selected)):
        if df_selected['trial'].iloc[i] != df_selected['trial'].iloc[i-1]:
            changed_rows.append(df_selected.iloc[i-1])
        if i == len(df_selected) - 1:
            changed_rows.append(df_selected.iloc[i])
    

    # Step 5: Create a new DataFrame from the collected rows
    df_changed = pd.DataFrame(changed_rows)
    
    # Step 6: Save the new DataFrame to a new CSV file
    df_changed.to_csv(output_csv, index=False)
```

```python
def calculate_success(row):
    if row['Xr'] < rect_pos[0] + rect_width and criterion_Yr * (-1) < row['Yr'] < criterion_Yr:
        return 1
    else:
        return 0

  


def model_trial_data(input_csv, output_csv):
    # Step 1: Read the input CSV file
    df = pd.read_csv(input_csv)
    
    # Step 2: Select specific columns
    selected_columns = ['Xr', 'Yr', 'date', 'trial', 'fish_id', 'session','ds','Yds','dus','duds']
    df_selected = df[selected_columns]
    
    # Step 3: Apply the calculate_success function to each row
    df_selected['ESuc'] = df_selected.apply(calculate_success, axis=1)

    # Step 4: Calculate the 'Edis' and 'EAng' columns
    df_selected['Edis'] = df_selected.apply(lambda row: math.sqrt((row['Xr'] - target_x)**2 + (row['Yr'] - target_y)**2), axis=1)
    df_selected['EAng'] = df_selected.apply(lambda row: math.atan2((row['Yr'] - target_y), (row['Xr'] - target_x)), axis=1)



    
    # Step 6: Save the new DataFrame to a new CSV file
    df_selected.to_csv(output_csv, index=False)
```

```python
def update_day_column(combined_df):
    # Ensure DataFrame is sorted correctly
    custom_order = ['Baseline', 'Rotation', 'Washout', 'ReRotation', 'ReWashout']
    combined_df['session'] = pd.Categorical(combined_df['session'], categories=custom_order, ordered=True)
    combined_df = combined_df.sort_values(by=['fish_id', 'session', 'date', 'trial']).reset_index(drop=True)
    
    # Initialize 'trial_id', 'day', and 'plotday' columns with 1
    combined_df['trial_id'] = 1
    combined_df['day'] = 1
    combined_df['plotday'] = 1

    # Update 'trial_id' based on date continuity
    for i in range(1, len(combined_df)):
        if combined_df['date'].iloc[i] == combined_df['date'].iloc[i-1]:
            combined_df.at[i, 'trial_id'] = combined_df.at[i-1, 'trial_id'] + 1
        else:
            combined_df.at[i, 'trial_id'] = 1

    # Update 'day' column based on session and date continuity
    for i in range(1, len(combined_df)):
        if (combined_df['date'].iloc[i] == combined_df['date'].iloc[i-1] and
            combined_df['session'].iloc[i] == combined_df['session'].iloc[i-1]):
            combined_df.at[i, 'day'] = combined_df.at[i-1, 'day']
        elif (combined_df['date'].iloc[i] != combined_df['date'].iloc[i-1] and
              combined_df['session'].iloc[i] == combined_df['session'].iloc[i-1]):
            combined_df.at[i, 'day'] = combined_df.at[i-1, 'day'] + 1
        elif (combined_df['session'].iloc[i] != combined_df['session'].iloc[i-1]):
            combined_df.at[i, 'day'] = 1

    # Update 'plotday' column based on fish_id and date continuity
    for i in range(1, len(combined_df)):
        if combined_df['fish_id'].iloc[i] != combined_df['fish_id'].iloc[i-1]:
            combined_df.at[i, 'plotday'] = 1
        elif (combined_df['fish_id'].iloc[i] == combined_df['fish_id'].iloc[i-1] and
              combined_df['date'].iloc[i] == combined_df['date'].iloc[i-1]):
            combined_df.at[i, 'plotday'] = combined_df.at[i-1, 'plotday']
        elif (combined_df['fish_id'].iloc[i] == combined_df['fish_id'].iloc[i-1] and
              combined_df['date'].iloc[i] != combined_df['date'].iloc[i-1]):
            combined_df.at[i, 'plotday'] = combined_df.at[i-1, 'plotday'] + 1

    return combined_df
```

```python
def calculate_session_csv(input_df):
    # Calculate median distances and median Yds for each plot day and fish ID
    median_distances = input_df.groupby(['fish_id', 'plotday'])['ds'].median().rename('mediandistance')
    median_Yds = input_df.groupby(['fish_id', 'plotday'])['Yds'].median().rename('medianYds')
    median_Ydsds = input_df.groupby(['fish_id', 'plotday'])['Ydsds'].median().rename('medianYdsds')
    median_dus = input_df.groupby(['fish_id', 'plotday'])['dus'].median().rename('mediandus')
    median_duds = input_df.groupby(['fish_id', 'plotday'])['duds'].median().rename('medianduds')
    median_Edis= input_df.groupby(['fish_id', 'plotday'])['Edis'].median().rename('medianEdis')
    median_EAng = input_df.groupby(['fish_id', 'plotday'])['EAng'].median().rename('medianEAng')

    mean_Yds = input_df.groupby(['fish_id', 'plotday'])['Yds'].mean().rename('meanYds')
    mean_distance = input_df.groupby(['fish_id', 'plotday'])['ds'].mean().rename('meandistance') 
    mean_dus = input_df.groupby(['fish_id', 'plotday'])['dus'].mean().rename('meandus')
    mean_duds = input_df.groupby(['fish_id', 'plotday'])['duds'].mean().rename('meanduds')
    mean_Edis = input_df.groupby(['fish_id', 'plotday'])['Edis'].mean().rename('meanEdis') 
    mean_EAng = input_df.groupby(['fish_id', 'plotday'])['EAng'].mean().rename('meanEAng')
    


    std_Edis = input_df.groupby(['fish_id', 'plotday'])['Edis'].std().rename('stdEdis')
    std_EAng = input_df.groupby(['fish_id', 'plotday'])['EAng'].std().rename('stdEAng')

    
    std_Xr = input_df.groupby(['fish_id', 'plotday'])['Xr'].std().rename('stdXr')
    std_Yr = input_df.groupby(['fish_id', 'plotday'])['Yr'].std().rename('stdYr')


    input_df['EAngd']= 0
    input_df['EAngd'] = np.degrees(input_df['EAng'])
    input_df['cos_mean'] = input_df['EAngd'].apply(lambda x: np.cos(x).mean())
    input_df['sin_mean'] = input_df['EAngd'].apply(lambda x: np.sin(x).mean())

    sum_cos_mean = input_df.groupby(['fish_id', 'plotday'])['cos_mean'].sum().rename('sum_cos')
    sum_sin_mean = input_df.groupby(['fish_id', 'plotday'])['sin_mean'].sum().rename('sum_sin')

    # Merge the median calculations back into input_df based on fish ID and plot day
    input_df = pd.merge(input_df, median_distances, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, median_Yds, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, median_Ydsds, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, median_dus, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, median_duds, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, median_Edis, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, median_EAng, how='left', on=['fish_id', 'plotday'])


    input_df = pd.merge(input_df, mean_distance, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, mean_Yds, how='left', on=['fish_id', 'plotday'])   
    input_df = pd.merge(input_df, mean_dus, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, mean_duds, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, mean_Edis, how='left', on=['fish_id', 'plotday'])   
    input_df = pd.merge(input_df, mean_EAng, how='left', on=['fish_id', 'plotday'])    


    input_df = pd.merge(input_df, std_Edis, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, std_EAng, how='left', on=['fish_id', 'plotday'])

    input_df = pd.merge(input_df, std_Xr, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, std_Yr, how='left', on=['fish_id', 'plotday'])
    
    input_df = pd.merge(input_df, sum_cos_mean, how='left', on=['fish_id', 'plotday'])
    input_df = pd.merge(input_df, sum_sin_mean, how='left', on=['fish_id', 'plotday'])



    # Ensure the data is sorted by fish_id, session, and plotday
    custom_order = ['Baseline', 'Rotation', 'Washout', 'ReRotation', 'ReWashout']
    input_df['session'] = pd.Categorical(input_df['session'], categories=custom_order, ordered=True)
    input_df = input_df.sort_values(by=['fish_id', 'session', 'plotday'])
    # Define a helper function for rolling calculations
    def calculate_moving_mean(df, col, window):
        df[f'moving_mean_{col}'] = df[col].rolling(window=window, min_periods=6).mean()
        return df
    def calculate_moving_median(df, col, window):
        df[f'moving_median_{col}'] = df[col].rolling(window=window, min_periods=6).median()
        return df

    # Apply rolling mean calculation for each group
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_mean(x, 'ds', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_mean(x, 'Yds', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_mean(x, 'dus', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_mean(x, 'duds', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_mean(x, 'Edis', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_mean(x, 'EAng', 12)).reset_index(drop=True)


    # Apply rolling median calculation for each group
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_median(x, 'ds', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_median(x, 'Yds', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_median(x, 'dus', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_median(x, 'duds', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_median(x, 'Edis', 12)).reset_index(drop=True)
    input_df = input_df.groupby(['fish_id', 'session']).apply(lambda x: calculate_moving_median(x, 'EAng', 12)).reset_index(drop=True)

    input_df['EST'] = 0
    for i in range(1, len(input_df)):
        if (input_df['plotday'].iloc[i] == input_df['plotday'].iloc[i-1] and
            input_df['ESuc'].iloc[i] == 1):
            input_df['EST'].iloc[i] =input_df['EST'].iloc[i-1] +1
        elif (input_df['plotday'].iloc[i] != input_df['plotday'].iloc[i-1] and
            input_df['ESuc'].iloc[i] == 1): 
            input_df['EST'].iloc[i] = 1    
        elif (input_df['plotday'].iloc[i] == input_df['plotday'].iloc[i-1] and
            input_df['ESuc'].iloc[i] == 0): 
            input_df['EST'].iloc[i] = input_df['EST'].iloc[i-1]
        elif (input_df['plotday'].iloc[i] != input_df['plotday'].iloc[i-1] and
            input_df['ESuc'].iloc[i] == 0):   
            input_df['EST'].iloc[i] = 0      


    output_file = new_df_path.replace('_trial_df.csv', '_session_df.csv')
    input_df.to_csv(output_file, index=False)

```
